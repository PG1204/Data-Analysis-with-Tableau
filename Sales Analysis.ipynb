```python
CostPerTransaction = data['NumberOfItemsPurchased'] * data['CostPerItem']
```

```python
SalesPerTransaction = data['NumberOfItemsPurchased'] * data['SellingPricePerItem']
```

```python
data["SalesPerTransaction"] = SalesPerTransaction
```

```python
data.head()
```

```python
ProfitPerTransaction = SalesPerTransaction - CostPerTransaction
```

```python
data['ProfitPerTransaction'] = ProfitPerTransaction
```

```python
Markup = ProfitPerTransaction / CostPerTransaction
```

```python
data['Markup'] = Markup
```

```python
data['Markup'] = round(data['Markup'], 2)
```
